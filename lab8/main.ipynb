```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1

trueS = 100
def f(x):
    return np.where(x < n, 10 * x / n, 10 * (x - 20) / (n - 20))

x = np.linspace(0, 20, 100)
a = 20 
b = 10 

N = 100

r_x = np.random.uniform(0, a, N)
r_y = np.random.uniform(0, b, N)

M = np.sum(r_y < f(r_x))

S = a * b * M / N

plt.plot(x, f(x), 'tab:orange')
plt.scatter(r_x, r_y)


plt.title(f"N = {N}, S = {S}, AE = {abs(S - trueS):.2f}, RE = {100 * abs(S - trueS) / trueS:.2f}")

plt.show()
```

```python
n = 1

trueVal = 16.17845

def f(x):
    return (11 - n * np.sin(x) ** 2) ** (0.5)

x = np.linspace(0, 5, 100)
a = 0
b = 5

N = 100

r_x = np.random.uniform(a, b, N)
r_y = f(r_x)

ans = (b - a) * np.mean(r_y)

plt.plot(x, f(x), 'tab:orange')
plt.scatter(r_x, r_y)


plt.title(f"N = {N}, Ans = {ans:.2f}, AE = {abs(ans - trueVal):.2f}, RE = {100 * abs(ans - trueVal) / trueVal:.2f}")

plt.show()
```

```python
theta = np.linspace(0, 2*np.pi, 100)

trueVal = np.pi

r = 1

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.plot(x, y, 'orange')

N = 1000

a, b = -r, r

r_x = np.random.uniform(a, b, N)
r_y = np.random.uniform(a, b, N)

plt.scatter(r_x, r_y)

f = lambda x, y: x ** 2 + y ** 2 < r ** 2

M = np.sum(f(r_x, r_y))

# S = 4 * r ** 2

ans = 4*M/N

plt.title(f"N = {N}, Ans = {ans:.2f}, AE = {abs(ans - trueVal):.2f}, RE = {100 * abs(ans - trueVal) / trueVal:.2f}")

plt.show()
```

```python
n = 1

A = 11 + n
B = 11 - n

theta = np.linspace(0, 2*np.pi, 100)

x = np.sqrt(A) * np.cos(theta)
y = np.sqrt(B) * np.sin(theta)


N = 1000

a, b = np.sqrt(A), np.sqrt(B)

r_x = np.random.uniform(-a, a, N)
r_y = np.random.uniform(-b, b, N)

phi = lambda x, y : np.arctan(y/x)
rad = lambda x, y : np.sqrt(x ** 2 + y ** 2)
p = lambda x, y : np.sqrt(A*np.cos(phi(x,y)) ** 2 + B*np.sin(phi(x,y)) ** 2)

plt.figure(figsize=(8, 8))
plt.plot(x, y, 'orange')
plt.scatter(r_x, r_y)

points =  np.array([(i, j) for i,j in zip(r_x, r_y) if rad(i,j) < p(i,j)])
# xs,  ys = zip(*points)
# plt.scatter(xs, ys)

S = a * b * 4

ans = S * len(points) / N

# f = lambda x, y: x ** 2 + y ** 2 < r ** 2
plt.title(f"N = {N}, Ans = {ans:.2f}")

plt.show()
```
